```python
from scipy.io import loadmat
import pandas as pd
import numpy as np
import os
from matplotlib import pyplot as plt
from matplotlib import image as mpimg
import cv2
```

```python
def visualize(img='001', segmentation=True, fundus_img=True):
    if segmentation:
        folder_path = 'C:\Luna_CS\Aravind\ORIGA\Semi-automatic-annotations'
        test_mat = f'{img}.mat'
        mat_file = os.path.join(folder_path, test_mat)
        mat = loadmat(mat_file)
        mask = mat['mask']
        plt.imshow(mask)
        plt.show()
    
    if fundus_img:
        imgs_folder = 'C:\Luna_CS\Aravind\ORIGA\Images'
        test_img = f'{img}.jpg'
        img_file = os.path.join(imgs_folder, test_img)
        img = mpimg.imread(img_file)
        plt.imshow(img)
```

```python
visualize(img='067')
```
